# Loss Functions

Implementation of common loss functions: Mean Squared Error (MSE) and Cross Entropy.

```python
import numpy as np

def mean_squared_error(y_true, y_pred):
    return np.mean(np.square(y_true - y_pred))

def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def categorical_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred)) / y_true.shape[0]

# Example Usage
y_true_reg = np.array([1.0, 2.0, 3.0])
y_pred_reg = np.array([1.1, 1.9, 3.2])
print(f"MSE: {mean_squared_error(y_true_reg, y_pred_reg)}")

y_true_bin = np.array([1, 0, 1, 1])
y_pred_bin = np.array([0.9, 0.1, 0.8, 0.99])
print(f"Binary Cross Entropy: {binary_cross_entropy(y_true_bin, y_pred_bin)}")
```
