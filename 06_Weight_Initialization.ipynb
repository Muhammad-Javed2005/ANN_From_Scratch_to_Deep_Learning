# Weight Initialization Techniques

Demonstrating different weight initialization methods using Keras.

```python
import tensorflow as tf
from tensorflow.keras import layers, models, initializers

def create_model(initializer):
    model = models.Sequential([
        layers.Dense(64, input_shape=(10,), kernel_initializer=initializer, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

initializers_list = {
    'Zeros': initializers.Zeros(),
    'RandomNormal': initializers.RandomNormal(mean=0.0, stddev=0.05),
    'GlorotNormal (Xavier)': initializers.GlorotNormal(),
    'HeNormal': initializers.HeNormal()
}

for name, init in initializers_list.items():
    print(f"--- Model with {name} Initialization ---")
    model = create_model(init)
    # Check weights of the first layer
    weights = model.layers[0].get_weights()[0]
    print(f"Mean: {weights.mean():.4f}, Std: {weights.std():.4f}\n")
```
