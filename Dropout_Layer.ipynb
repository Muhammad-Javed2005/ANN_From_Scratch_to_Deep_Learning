```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout

```

```python
data = pd.read_csv(r"E:\AI and Data Science\DL with Engr.Muhammad Javed\Dataset\Churn_Modelling.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['RowNumber','CustomerId','Surname' , "Geography" , "Gender" ],inplace=True)
```

```python
data
```

```python
input_data = data.iloc[: , :-1]
output_data = data.iloc[:,-1]

```

```python
input_data
```

```python
output_data
```

```python
input_data.shape
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(input_data , output_data , test_size=0.2 , random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data) , columns=input_data.columns)
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6 , input_dim= 8 , activation="relu" , kernel_regularizer=l2(l2=0.01)))
ann.add(Dropout(0.5))
ann.add(Dense(4 , activation = "relu" , kernel_regularizer=l2(l2=0.01)))
ann.add(Dropout(0.5))
ann.add(Dense(2 , activation = "relu" , kernel_regularizer=l2(l2=0.01)))
ann.add(Dropout(0.5))
ann.add(Dense(1 , activation  = "sigmoid" ))
```

```python
ann.compile(optimizer="Adam" , loss = "binary_crossentropy" , metrics=  ["accuracy"] )
```

```python
ann.fit(X_train , Y_train , batch_size=100 , epochs=50 , validation_data=(X_test , Y_test)  , callbacks=EarlyStopping()  )
```

```python
prd = ann.predict(X_test)
prd_data = []
for i in prd:
    if i>=0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
prd_data
```

```python
prd1 = ann.predict(X_train)
prd_data1 = []
for i in prd1:
    if i>=0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
prd_data1
```

```python
accuracy_score(Y_test , prd_data)*100
```

```python
accuracy_score(Y_train , prd_data1)*100
```

```python
ann.history.history.keys()
```

```python
training_acc = ann.history.history['accuracy']
testing_acc = ann.history.history['val_accuracy']
```

```python
plt.plot([i for i in range(len(training_acc))] , training_acc , color='blue' , label='Training Accuracy')
plt.plot([i for i in range(len(testing_acc)) ], testing_acc , color='red' , label='Testing Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
